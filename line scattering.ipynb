```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load Data
df = pd.DataFrame({
    "hours": [1, 2, 3, 4, 5, 6, 7],
    "score": [45, 55, 65, 72, 80, 88, 95]
})

# 2. Plot with sns.relplot()
g = sns.relplot(data=df, x="hours", y="score", kind="scatter")

# 3. Add Title
g.fig.suptitle("Study Hours vs Exam Score", y=1.03)

plt.show()
```

There is a strong positive linear relationship. As the number of study hours increases, the exam score increases steadily.

```python
df = pd.DataFrame({
    "hours": [1,2,3,4,5,6,7,8],
    "score": [50,60,68,75,80,85,92,98],
    "gender": ["Male","Male","Male","Male","Female","Female","Female","Female"]
})

# 2. Plot with hue and style mapping
g = sns.relplot(data=df, x="hours", y="score", hue="gender", style="gender", kind="scatter")

# 3. Add Title
g.fig.suptitle("Exam Scores by Gender", y=1.03)

plt.show()
```

How are males and females distinguished? They are distinguished by two visual identifiers simultaneously: color (different hues) and marker shape (e.g., a circle vs. an 'X' or square).
What does hue represent? hue maps a categorical variable to colors, letting you group data points by distinct subgroups.
What does style represent? style maps a categorical variable to marker shapes, adding a secondary visual.

```python
df= pd.DataFrame({
    "hours": [1,2,3,4,5,6],
    "score": [48,58,67,74,83,90],
    "assignments": [2,4,3,6,5,8]
})

# 2. Plot with size mapping
graph = sns.relplot(data=df, x="hours", y="score", size="assignments", sizes=(20, 200), kind="scatter")

# 3. Add Title
graph.fig.suptitle("Effect of Assignments on Scores", y=1.03)

plt.show()
```

Which students completed the most assignments? The student who studied for 6 hours and scored a 90 completed the most assignments (8 assignments, indicated by the largest circle on the plot).
What does the size of a point represent? It represents a quantitative (numeric) value. Points scale dynamically: smaller circles represent fewer completed assignments, while larger circles represent more completed assignments.

```python
df = pd.DataFrame({
    "hours": [1,2,3,4,1,2,3,4],
    "score": [50,60,70,80,55,65,75,85],
    "department": ["CS","CS","CS","CS","IS","IS","IS","IS"]
})

# 2. Plot using column faceting (col)
g = sns.relplot(data=df, x="hours", y="score", col="department", kind="scatter")

# 3. Add Title
g.fig.suptitle("Department Comparison", y=1.05)

plt.show()
```

Which department appears to have higher scores? The IS (Information Systems) department has higher scores for the exact same amount of study hours (e.g., at 1 hour, IS scored 55 while CS scored 50).
What does the col argument do? It enables faceting (subplots). It splits your dataset based on a category and draws separate graphs side-by-side in columns so you can compare categories clearly without cluttering a single plot

```python
df = pd.DataFrame({
    "experience": [1,2,3,4,5,6,7,8],
    "salary": [30000,35000,42000,50000,58000,65000,72000,80000]
})

# 2. Plot with regression model tracking
g = sns.lmplot(data=df, x="experience", y="salary")

# 3. Add Title
g.fig.suptitle("Experience vs Salary", y=1.03)

plt.show()
```

1 What does the regression line indicate? It indicates the overall trend or line of best fit. It shows a positive linear correlation, confirming that every additional year of experience directly leads to a steady, predictable increase in salary.If someone has 9 years of experience, approximately
2 what salary would you predict? Following the trajectory of the line, at $x = 9$, the predicted salary would be approximately $87,000 to $88,000.
3 What does the gray shaded area around the line represent? It represents the 95% Confidence Interval. This shaded boundary shows the statistical margin of error or uncertainty around the regression estimate line. (Because this data is almost perfectly linear, the band will look narrow).
