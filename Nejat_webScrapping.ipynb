```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = "https://www.fanabc.com/english/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
articles = soup.find_all("article")
if not articles:
    articles = soup.find_all("div", class_="post-inner")
print(f"{'CATEGORY'} | {'ARTICLE TITLE'}")
print("-" * 80)
for article in articles:
    title_tag = article.find("h3") or article.find("h2") or article.find("a")
    if title_tag:
        title = title_tag.get_text(strip=True)
    else:
        continue
    cat_tag = article.find("a", class_="post-cat") or article.find("span", class_="meta-cat")
    category = cat_tag.get_text(strip=True) if cat_tag else "General"
    print(f"{category} | {title}")
```

```python
file=open("fanabc_news.csv", "w", encoding="utf-8", newline="")
writer = csv.writer(file)

    # Write the headers (columns)
writer.writerow(["Title", "Link", "Category"])

    # 4. Loop through each article block and extract the details
for article in articles:
        # Extract the Title and Link
        title_tag = article.find("h3") or article.find("h2") or article.find("a")

        if title_tag:
            link_tag = title_tag.find("a")
            title = link_tag.get_text(strip=True)
            link = link_tag.get("href")
        elif title_tag:
            title = title_tag.get_text(strip=True)
            link = url  # Fallback if specific link isn't parsed
        else:
            continue

        # Extract Category (if available in the post metadata)
        category_tag = article.find("a", class_="post-cat") or article.find(
            "span", class_="meta-cat"
        )
        category = (
            category_tag.get_text(strip=True) if category_tag else "General News"
        )

        # Write the row data to our CSV file
        writer.writerow([title, link, category])

print("Data successfully extracted and saved to fanabc_news.csv!")
```
