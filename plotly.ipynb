```python
import numpy as np
departments = ('CS', 'IS', 'SE', 'IT')
students = {
    'Male': np.array([70, 65, 80, 50]),
    'Female': np.array([60, 55, 70, 45])
}

# Initialize subplots
fig, ax = plt.subplots()
width = 0.7

# Plot base bars (Male)
male_bars = ax.bar(departments, students['Male'], width=width, label='Male', color='#3498db')

# Plot stacked bars (Female) using the bottom parameter
female_bars = ax.bar(departments, students['Female'], width=width, bottom=students['Male'], label='Female', color='#e74c3c')

# Add values inside each section of the bars
ax.bar_label(male_bars, label_type='center', color='white', fontweight='bold')
ax.bar_label(female_bars, label_type='center', color='white', fontweight='bold')

# Formatting
ax.set_title('Students by Department and Gender')
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------
# 1. CONSOLIDATED DATASET REGISTRY
# ----------------------------------------------------
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 150, 170, 160, 190]

fruits = ['Apple', 'Orange', 'Banana', 'Mango']
counts = [50, 70, 40, 60]
fruit_colors = ['#ff7675', '#fab1a0', '#ffeaa7', '#55efc4']

movies = ['The Shining', 'Star Wars', 'Inception', 'Gladiator']
ratings = [8.4, 8.6, 8.8, 8.5]

departments = ('CS', 'IS', 'SE', 'IT')
gender_data = {
    'Male': np.array([70, 65, 80, 50]),
    'Female': np.array([60, 55, 70, 45])
}

# ----------------------------------------------------
# 2. INITIALIZE 2x2 DASHBOARD MATRIX
# ----------------------------------------------------
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# ----------------------------------------------------
# PLOT 1: Monthly Sales Trend (Top-Left Subplot)
# ----------------------------------------------------
axs[0, 0].plot(months, sales, color='green', linestyle='--', marker='s', markersize=6, label='Sales')
axs[0, 0].set_title('Monthly Sales Trend', fontsize=12, fontweight='bold')
axs[0, 0].legend()
axs[0, 0].grid(axis='y', linestyle=':', alpha=0.6)

# ----------------------------------------------------
# PLOT 2: Fruit Supply (Top-Right Subplot)
# ----------------------------------------------------
axs[0, 1].bar(fruits, counts, color=fruit_colors)
axs[0, 1].set_title('Fruit Supply Inventory', fontsize=12, fontweight='bold')
axs[0, 1].set_ylabel('Quantity')

# ----------------------------------------------------
# PLOT 3: Movie Ratings (Bottom-Left Subplot)
# ----------------------------------------------------
axs[1, 0].barh(movies, ratings, color='#9b59b6')
axs[1, 0].set_title('Movie Ratings Evaluation', fontsize=12, fontweight='bold')
axs[1, 0].set_xlabel('Score Index')
axs[1, 0].set_xlim(0, 10)  # Standardize score scale limit
axs[1, 0].grid(axis='x', linestyle=':', alpha=0.6)

# ----------------------------------------------------
# PLOT 4: Student Distribution by Gender (Bottom-Right Subplot)
# ----------------------------------------------------
m_bars = axs[1, 1].bar(departments, gender_data['Male'], width=0.6, label='Male', color='#34495e')
f_bars = axs[1, 1].bar(departments, gender_data['Female'], width=0.6, bottom=gender_data['Male'], label='Female', color='#1abc9c')

# Attach values labels directly onto sections
axs[1, 1].bar_label(m_bars, label_type='center', color='white', fontweight='bold')
axs[1, 1].bar_label(f_bars, label_type='center', color='white', fontweight='bold')

axs[1, 1].set_title('Student Distribution by Gender', fontsize=12, fontweight='bold')
axs[1, 1].legend()

# ----------------------------------------------------
# 3. GLOBAL MATRIX LAYOUT REFINEMENT
# ----------------------------------------------------
# Suppress overlapping labels across subplots automatically
plt.tight_layout()

# Display the dashboard output
plt.show()
```
