```python
#Task 2: Number of Movies Released Each Year
movies_year = movies.groupby("year").size()
plt.figure(figsize=(10, 5))
plt.plot(
    movies_year.index,
    movies_year.values,
    marker="s",
    color="g",
    linestyle="--",label='Number of movie'
)
plt.title("Number of Movies Released Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.legend()
plt.show()
```

```python
#Task 3: Top 10 Movie Genres
top_genres = movies["genre"].value_counts().head(10).sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(top_genres.index, top_genres.values, color="blue", edgecolor="black")
ax.set_xlabel("Genre")
ax.set_ylabel("Number of Movies")
ax.set_title("Top 10 Movie Genres")
plt.show()
```

```python
1. Which genre contains the most movies? comedy
2. Which genre contains the fewest movies among the top 10? mystery
3. Why is a bar chart suitable for this task?Bar charts are ideal for comparing discrete not continous  
genres with regard to quantitative count.
```

```python
#Task 4: Top 10 Directors
top_directors = (
    movies["director"].value_counts().head(10).sort_values(ascending=True))

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(top_directors.index, top_directors.values, color="green", edgecolor="black")
ax.set_xlabel("Number of Movies")
ax.set_ylabel("Director")
ax.set_title("Top 10 Directors by Number of Movies")

plt.tight_layout()
plt.show()
```

```python
1,Which director produced the most movies?woody allen
2. Why is a horizontal bar chart more suitable than a vertical bar chart for this visualization?
  Long string names (like director names) are much easier to read horizontally on the y-axis without overlapping or 
  needing steep text rotation.
```

```python
#Task 5: Distribution of Movie Ratings
plt.figure(figsize=(10, 5))
plt.hist(
    movies["score"].dropna(),
    bins=20,
    color="gold",
    edgecolor="black",
    alpha=0.7,
)
plt.title("Distribution of Movie Ratings (Scores)")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
#Task 6: Average Movie Score by Genre
avg_score_genre = movies.groupby("genre")["score"].mean().sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 6))
avg_score_genre.plot(kind="bar", color="orange", edgecolor="black")
plt.title("Average Movie Score by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Score")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Task 7: Top 10 Highest Grossing Movies
top_grossing = movies.nlargest(10, "gross")[["name", "gross"]]
plt.figure(figsize=(10, 6))
plt.barh(
    top_grossing["name"][::-1],
    top_grossing["gross"][::-1] / 1e9,
    color="purple",
    edgecolor="black",
)
plt.title("Top 10 Highest Grossing Movies")
plt.xlabel("Gross Revenue (in Billions USD)")
plt.ylabel("Movie Title")
plt.tight_layout()
plt.show()
```
